# **3択暗記システム**

## **概要**

3択クイズができるプログラムです。
自分でクイズを登録できます。

3つのモードがあります。

1つ目は、"問題を登録する"モードです。自分が解きたい問題を登録することができます。

2つ目は、"覚える"モードです。登録されている問題を一覧で表示できます。答えはオレンジで表示されるため、赤シートで隠すことができます。

3つ目は、"テスト"モードです。
登録した問題が3択クイズで遊べます。
出題する問題数、出題の順番などを設定できます。
また、問題ごとに正解/不正解、最後には正解した問題数や正答率が表示されます。

## **操作方法**
1つ目の"問題を登録する"モードでは、5つの項目を入力してください。

入力する項目は、問題文/選択肢①/選択肢②/選択肢③/答え の5つです。

"問題を登録する"ボタンを押してください。入力ボックスの上に書かれている指示を見ながら、それぞれの項目について入力し、送信ボタンを押してください。

答え以外の4項目は文章、単語、数値などどんな形でも構いません。

その後、自らが入力した選択肢が番号とともに表示されます。そこから答えと対応する半角数字を入力してください。半角数字以外を入力した場合、再入力となります。答えとなる文字列では入力できません。



## **制作者のコメント**

作り始めた当初は、早押しクイズが楽しめるプログラムを作ろうと考えていました。
しかし、Pythonがボタンを押した際の素早い反応などに優れておらず、早押しクイズの特性と合っていませんでした。
そのことから、今回の暗記システム制作に変更しました。

今回のプログラムで工夫したポイントはモード選択画面です。
ボタンの上にカーソルを置いて少し経つとモードの説明が表示されるようにしました。
また、テストモードでランダムに問題を出力する際には、同じ問題が2度出題されないように工夫しました。

苦労したポイントは、テストモードで問題を出力させる所です。問題の違う出題順番にするためにif文を用い、全て同じ関数で動かせるようにしました。

ボタンなどWidgetを多く使用したため、コードがごちゃついてしまったところは反省点です。

```python
# ライブラリのインストール
from IPython.display import clear_output
import time
import ipywidgets as widgets
from IPython.display import display
import re
import random
```

```python
#リスト
question_list=[]
selection_list_1=[]
selection_list_2=[]
selection_list_3=[]
answer_list=[]

# 入力システム
def text_input_system(input_sentence,text_input_system_callback):
  print(input_sentence)
  text_input_button = widgets.Text(
    description="",
    placeholder="ここに入力")

  send_button = widgets.Button(
    description="送信",
    layout=widgets.Layout(width="50px"),
    button_style="success")
  text_send_button = widgets.HBox([text_input_button, send_button])
  display(text_send_button)

  def send_text(button):
    text_input_button_value = text_input_button.value.strip()
    text_input_button.value = ""
    text_send_button.close()
    text_input_system_callback(text_input_button_value)

  send_button.on_click(send_text)

#back
def back():
  back_button = widgets.Button(description="←戻る",
                               layout=widgets.Layout(width="70px"),
                               button_style="primary")
  back_button.on_click(show_screen_home)
  display(back_button)


# ホーム画面
def show_screen_home(button=None):
  clear_output()
  print("メニューを選んでください")
  print()
  #ボタンの動作
  #・問題の登録
  register_button = widgets.Button(description="問題を登録する",
                                   button_style="success",
                                   tooltip="覚えたい問題を登録できます")
  register_button.on_click(show_screen_add)
  display(register_button)

  #・問題の一覧表示
  memory_button = widgets.Button(description="覚える",
                                 layout=widgets.Layout(margin="15px 0px"),
                                 button_style="info",
                                 tooltip="登録した問題が一覧で表示されます")
  memory_button.on_click(show_screen_memory)

  #・テスト
  test_button = widgets.Button(description="テスト",
                               layout=widgets.Layout(margin="15px 15px"),
                               button_style="info",
                               tooltip="登録した問題に正誤判定ありで答えられます")
  test_button.on_click(show_screen_test_setting)
  button_width = widgets.HBox([memory_button, test_button])
  display(button_width)

# 問題文登録画面
def show_screen_add(button=None):
  clear_output()
  back()
  text_input_system("問題文を入力してください(入力なしで戻る)",add_question1)

def add_question1(add_question_input):
  clear_output()
  if add_question_input == "":
    show_screen_home()
  else:
    question_list.append(add_question_input)
    text_input_system("1つ目の選択肢を入力してください",add_question2)

def add_question2(add_question_input):
  clear_output()
  selection_list_1.append(add_question_input)
  text_input_system("2つ目の選択肢を入力してください", add_question3)

def add_question3(add_question_input):
  clear_output()
  selection_list_2.append(add_question_input)
  text_input_system("3つ目の選択肢を入力してください", add_question4)

def add_question4(add_question_input):
  clear_output()
  selection_list_3.append(add_question_input)
  question_number = len(question_list) - 1
  print("1. " + selection_list_1[question_number])
  print("2. " + selection_list_2[question_number])
  print("3. " + selection_list_3[question_number])
  text_input_system("答えになる選択肢の\033[31m番号を半角\033[0mで入力してください", add_question5)

def add_question5(add_question_input):
  clear_output()
  if re.fullmatch(r"[1-3]", add_question_input):
    print("\033[33m問題が登録されました！\033[0m")
    answer_list.append(int(add_question_input))
    time.sleep(2)
    clear_output()
    show_screen_home()
  else:
    text_input_system("答えになる選択肢を\033[31m1～3の半角数字\033[0mで入力してください", add_question5)

#覚える画面
def show_screen_memory(button=None):
  clear_output()
  back()
  for i in range(len(question_list)):
    print(f"Q{i+1}. {question_list[i]}")
    if answer_list[i]==1:
      print(f"――\033[33m{selection_list_1[i]}\033[0m")
    elif answer_list[i]==2:
      print(f"――\033[33m{selection_list_2[i]}\033[0m")
    elif answer_list[i]==3:
      print(f"――\033[33m{selection_list_3[i]}\033[0m")
    else:
      print("-----")

#テスト画面_設定
def show_screen_test_setting(button=None):
  clear_output()
  back()
  print("問題の出題数、出題方法を選んでください")
  #出題数スライダー
  test_slider = widgets.IntSlider(
    value=5,
    min=1,
    max=50,
    step=1,
    description="出題数："
    )
  display(test_slider)
  print(f"{'　'*6}登録されている問題が出題数よりも少ない場合は、\n{'　'*6}登録されている問題がすべて出力されると終了になります")
  print()
  #ランダムON/OFF
  test_radio = widgets.RadioButtons(
    options=["ランダムに出題する", "登録が古い順に出題する", "登録が新しい順に出題する"],
    description="出題順：",
    )
  display(test_radio)
  #決定ボタン
  test_setting_button = widgets.Button(
    description="テストを始める",
    layout=widgets.Layout(margin="0px 0px 0px 70px"),
    button_style="success")
  display(test_setting_button)
  #値保存
  def on_click_test_setting(button):
    question_count = test_slider.value
    testmode = test_radio.value
    test(question_count, testmode)

  test_setting_button.on_click(on_click_test_setting)


#テストモード
def test(question_count,testmode):
  correct=0
  wrong=0
  i=0
  n=-1
  n_random=[]
  n_random = list(range(len(question_list)))
  random.shuffle(n_random)

  #出題
  def show_question(i):
    #i==i問目
    #n==問題番号
    nonlocal correct, wrong
    question_count_2 = min(question_count,len(question_list))
    clear_output()
    back()
    nonlocal n
    i=i+1
    if testmode == "ランダムに出題する":
      n=n_random[i-1]
    elif testmode =="登録が古い順に出題する":
      n=n+1
    elif testmode == "登録が新しい順に出題する":
      n=len(question_list)-i
    print(n)
    print(f"{i}/{question_count_2}問目：")
    print("Q. "+question_list[n])

    answer_radio = widgets.RadioButtons(
    options=[selection_list_1[n],selection_list_2[n],selection_list_3[n]],
    description="解答",
    )
    submit_button = widgets.Button(description="解答する", button_style='primary')
    display(answer_radio, submit_button)
  #正誤判定
    def question_judge(_):
      nonlocal correct, wrong
      if answer_radio.index==0:
        user_answer = selection_list_1[n]
      elif answer_radio.index==1:
        user_answer = selection_list_2[n]
      elif answer_radio.index==2:
        user_answer = selection_list_3[n]


      if answer_list[n]==1:
        correct_answer = selection_list_1[n]
      elif answer_list[n]==2:
        correct_answer = selection_list_2[n]
      elif answer_list[n]==3:
        correct_answer = selection_list_3[n]

      if user_answer==correct_answer:
        print("\033[31m〇正解です！〇\033[0m")
        correct+=1
      else:
        print("\033[34m×不正解です！×\033[0m")
        wrong+=1
      print(f"あなたの答え: {user_answer}")
      print(f"\033[31m正解: {correct_answer}\033[0m")

      if i < question_count_2:
        next_question_button = widgets.Button(
          description="次へ",
          layout=widgets.Layout(margin="0px 0px 0px 70px"),
          button_style="success")
        display(next_question_button)
        def next_question(button):
          show_question(i)
        next_question_button.on_click(next_question)

      else:
        next_question_button = widgets.Button(
          description="次へ",
          layout=widgets.Layout(margin="0px 0px 0px 70px"),
          button_style="success")
        display(next_question_button)

        def show_end_screen(button):
          clear_output()
          back()
          print("\033[47m==結果発表==\033[0m")
          print("お疲れさまでした！")
          print(f"{question_count_2}問中、{correct}問正解！")
          if correct==0:
            print("正答率は0%でした！")
          else:
            print(f"正答率{int((correct/question_count_2)*100)}%！")
          test_setting_button = widgets.Button(description="もう一度テスト",
                                               layout=widgets.Layout(margin="5px 10px"),
                                               button_style="success")
          display(test_setting_button)
          back_button = widgets.Button(description="メニュー画面へ",
                                       layout=widgets.Layout(margin="5px 10px"),
                                       button_style="primary")
          back_button.on_click(show_screen_home)
          test_setting_button.on_click(show_screen_test_setting)
          display(back_button)

        def next_question(button):
          show_end_screen()
        next_question_button.on_click(show_end_screen)
    submit_button.on_click(question_judge)

  if len(question_list)==0:
    clear_output()
    print("問題を登録してください")
    back_button = widgets.Button(description="メニュー画面へ",
                                 layout=widgets.Layout(margin="5px 10px"),
                                 button_style="primary")
    display(back_button)
    back_button.on_click(show_screen_home)
  else:
    show_question(0)

#初期動作
show_screen_home()
```
